```python
import os
import shutil
import random

# ===== PARAMÈTRES =====
SOURCE_DIR = r"C:\Users\Bureau\Desktop\ecg_pipeline\data\inference_inputs"
TARGET_DIR = r"C:\Users\Bureau\Desktop\ecg_pipeline\data\dataset"
TRAIN_RATIO = 0.8
SEED = 42
EXTENSIONS = (".png", ".jpg", ".jpeg")

# =====================
random.seed(SEED)

train_dir = os.path.join(TARGET_DIR, "train")
test_dir = os.path.join(TARGET_DIR, "test")
os.makedirs(train_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)

# Collecter toutes les images
all_images = []

for patient_id in os.listdir(SOURCE_DIR):
    patient_path = os.path.join(SOURCE_DIR, patient_id)
    if os.path.isdir(patient_path):
        for file in os.listdir(patient_path):
            if file.lower().endswith(EXTENSIONS):
                all_images.append(os.path.join(patient_path, file))

# Mélange
random.shuffle(all_images)

# Split
split_idx = int(len(all_images) * TRAIN_RATIO)
train_images = all_images[:split_idx]
test_images = all_images[split_idx:]

def copy_images(image_list, destination):
    for img_path in image_list:
        filename = os.path.basename(img_path)
        dst = os.path.join(destination, filename)
        shutil.copy2(img_path, dst)

copy_images(train_images, train_dir)
copy_images(test_images, test_dir)

print(f"✔ Images train : {len(train_images)}")
print(f"✔ Images test  : {len(test_images)}")
```
